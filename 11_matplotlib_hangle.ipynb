```python
import warnings
warnings.filterwarnings(action='ignore')
%config Computer.use_jedi = False
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib as mpl
import matplotlib.font_manager as fm
# 한글 폰트를 사용할 때 마이너스 부호가 깨지는 문제를 대처한다. 
mpl.rcParams['axes.unicode_minus'] = False
```

```python
data = np.random.randint(-100, 100, 50).cumsum()
data
```

```python
plt.plot(data)
plt.show()
```

```python
plt.figure(figsize=(4,2)) # 차트 크기 설정 
plt.plot(range(50, 100), data, 'r')
plt.title('시간별 가격 추이') # 차트 제목
plt.xlabel('시간(분)') # x축 제목
plt.ylabel('시간(분)') # y축 제목
plt.show()
```

한글 폰트를 사용하는 3가지 방법 

<1> FontProperties 사용 => 그래프에 폰트가 필요한 항목마다 개별적으로 지정한다. 

```python
# 텍스트를 지정하는 항목에 지정해서 사용할 수 있고 지정해준 항목에만 해당 폰트가 적용된다. 

font = 'C:\\Windows\\Fonts\\NanumGothicCoding.ttf'
fontprop = fm.FontProperties(fname=font)
```

```python
plt.figure(figsize=(4,2))
plt.plot(range(50, 100), data, 'r')
plt.title('시간별 가격 추이', fontproperties=fontprop) 
plt.xlabel('시간(분)', fontproperties=fontprop) 
plt.ylabel('시간(분)', fontproperties=fontprop)
plt.show()
```

<2> matplotlib 라이브러리의 rcParams[]로 전역 글꼴 설정 => 그래프의 폰트 설정이 필요한 파일마다 지정한다. 

```python
# matplotlib 폰트 기본 설정 정보 읽기 
print('기본으로 설정된 폰트 이름:', plt.rcParams['font.family'])
print('기본으로 설정된 폰트 크기:', plt.rcParams['font.size'])
```
