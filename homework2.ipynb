```python
import requests
from bs4 import BeautifulSoup

url = 'https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&query=%EC%BF%A8+%EC%9D%B4%EC%97%AC%EB%A6%84++%EA%B0%80%EC%82%AC&oquery=%EC%BF%A8+%EC%9D%B4%EC%97%AC%EB%A6%84+%EB%A8%B8%EB%93%9C+%EA%B0%80%EC%82%AC&tqi=iM1ehlprvxsssRABmONssssstPK-063671'

r = requests.get(url)

r.status_code # HTTP 응답코드
```

```python
soup = BeautifulSoup(r.text, 'html.parser')

print(soup.prettify())
```
